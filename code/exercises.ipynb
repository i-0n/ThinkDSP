```python
import thinkdsp
```

```python
my_wave = thinkdsp.read_wave('72475__rockwehrmann__glissup02.wav')
```

```python
segment = my_wave.segment(start=0.4, duration=0.9)
```

```python
segment.plot()
```

```python
segment.make_audio()
```

```python
spectrum = segment.make_spectrum()
```

```python
spectrum.plot()
```

```python
spectrum.low_pass(4500)
```

```python
spectrum.plot()
```

```python
filtered = spectrum.make_wave()
```

```python
filtered.make_audio()
```

```python
cos = thinkdsp.CosSignal(freq=220, amp=1, offset=0)
sin = thinkdsp.SinSignal(freq=330, amp=0.8, offset=0)
```

```python
mixed = cos + sin
```

```python
mixed.plot()
```

```python
wave = mixed.make_wave()
```

```python
spec = wave.make_spectrum()
```

```python
spec.plot()
```

```python
mixed.period
```

```python
spec_1 = wave_1.make_spectrum()
```
